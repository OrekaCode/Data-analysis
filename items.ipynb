```python
''' Using  Numpy, Pandas and Matplot libratries to analysis this csv file '''

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("items.csv")  # Reading the csv_file
df.head() 
```
